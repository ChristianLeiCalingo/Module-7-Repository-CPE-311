1. Using web scraping, search for the list of the hospitals, their address and contact information. Save the list in a new csv file, hospital.csv


```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_hospitals_in_the_Philippines?fbclid=IwAR3tlpYW_R1fYSs4u1BigxwEM2aSjoPBcQcN7wbd-_u-hbrcmXwRZNVCJhU'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')
```

```python
table = soup.find('table', class_ = 'wikitable')
```

```python
hospitals = table.find_all('th')
hospitals
print(hospitals)
```

```python
hospitals= pd.DataFrame(columns = [title.text.strip() for title in hospitals])
hospitals
```

```python
column_data = table.find_all('tr')
print(column_data)
```

```python
for row in column_data:
  row_data = row.find_all('td')
  individual_row = [data.text.strip() for data in row_data]
  print(individual_row)
```

```python
for row in table.find_all('tr')[1:]:
  data = row.find_all('td')
  results = [datas.text.strip() for datas in data]
  hospitals.loc[len(hospitals)] = results
```

```python
hospitals
```

```python
hospitals.to_csv('data/hospitals.csv', index = False)
```
